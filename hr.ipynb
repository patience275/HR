## feature eng

```python

total_working_years_binned = pd.cut(df['TotalWorkingYears'], bins=range(0, df['TotalWorkingYears'].max() + 3, 2), right=False, labels=False)
peer_group_salary = df.groupby(['JobRole', total_working_years_binned])['MonthlyIncome'].transform('mean')
df['Salary_Peer_Gap'] = ((df['MonthlyIncome'] - peer_group_salary) / peer_group_salary) * 100
```

```python

Overtime_Score = df['OverTime'].apply(lambda x: 2 if x == 1 else 0)

Commute_Score = df['DistanceFromHome'] / 15

WorkLife_Score = df['WorkLifeBalance'].apply(lambda x: 4 - x)

def calculate_travel_score(travel):
    if travel == 'Non-Travel':
        return 0
    elif travel == 'Travel_Rarely':
        return 1
    elif travel == 'Travel_Frequently':
        return 2
    else:
        return 0

Travel_Score = df['BusinessTravel'].apply(calculate_travel_score)


df['Burnout_Risk_Score'] = (Overtime_Score+ Commute_Score +
                            WorkLife_Score + Travel_Score) / 8 * 100

display(df['Burnout_Risk_Score'].head())
```

##kmeans

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow method (optional, but good practice)
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

# Visualize elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
for i in range(2, 11): # Silhouette score is not defined for a single cluster
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(scaled_df)
    silhouette_scores.append(silhouette_score(scaled_df, kmeans.labels_))


print("\nSilhouette Scores:")
for i, score in enumerate(silhouette_scores):
    print(f"Silhouette Score for {i+2} clusters: {score}")
```

```python
kmeans_df=df.copy()
```

```python
kmeans=KMeans(n_clusters=4,random_state=42)
kmeans.fit(scaled_df)
labels=kmeans.predict(scaled_df)
kmeans_df['clusters']=labels
```

```python
score = silhouette_score(scaled_df, labels)
print(f' Cluster={len(set(labels))}, Silhouette={score:.3f}')
```

```python
kmeans_df['clusters'].unique()
```

```python
numeric_cols = kmeans_df.select_dtypes(include=['number']).columns
cluster_summary = kmeans_df.groupby('clusters')[numeric_cols].mean()
display(cluster_summary)
```

## hirechial clustering

```python
hier_df=df.copy()
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
import pandas as pd

plt.figure(figsize=(10, 6))
Z = sch.linkage(scaled_df, method='ward')
dendrogram = sch.dendrogram(Z)
plt.title("Dendrogram (Hierarchical Clustering)")
plt.xlabel("Employees")
plt.ylabel("Distance")
plt.show()

# Cut the tree at distance threshold = 20 and add to dataset
labels = fcluster(Z, t=20, criterion='distance')
hier_df['Cluster'] = labels # Assuming you want to add the cluster labels to the original df

print("\nCluster distribution based on distance threshold 20:")
print(hier_df['Cluster'].value_counts())
```

```python
for dist in range(10,25):
    labels = fcluster(Z, t=dist, criterion='distance')
    score = silhouette_score(scaled_df, labels)
    print(f"Distance={dist}, Cluster={len(set(labels))}, Silhouette={score:.3f}")
```

```python
numeric_cols = hier_df.select_dtypes(include=['number']).columns
cluster_summary = hier_df.groupby('Cluster')[numeric_cols].mean()
display(cluster_summary)
```
